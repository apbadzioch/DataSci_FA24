## Part 1: Types of Graphs and Plots

#### 1. Line Plots
- one of the simplest plots
- they indicate the progress of a trend over time

```python
# let's create the dataset
a = np.linspace(0, 10)

# creating the line plot with matplotlib.pyplot
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
```

```python
# creating the dataset
x = pd.Series(['sunday', 'monday', 'tuesday', 'wednesday', 'thursday'])
y = pd.Series(['running', 'tennis', 'football', 'biking', 'ping-pong'])

# creating the plot
plt.plot(x, y)

# adding more items to the plot
plt.title('days vs sports') # adding a title to the graph
plt.xlabel('days') # adding x label
plt.ylabel('sports') # adding y label

# adding a grid
plt.grid()
```

#### 2. Bar Plots
- bar plots are used to compare different categories

```python
# creating the dataset
df1 = pd.DataFrame.from_dict(
    {'year': [2016, 2017, 2018, 2022],
     'men': [300, 305, 325, 350],
     'women': [280, 301, 325, 355],
     'total': [580, 606, 650, 705]
     }
)

# creating the bar plot
plt.bar(x=df1['year'], height=df1['total'])
```

#### 3. Pie Charts
- pie charts are used to indicate the proportion over the whole

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# creating the plot
plt.pie(y);
```

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# adding labels
cars = ['benz', 'honda', 'ford', 'nissan']

# making the wedges stand out
wedge = [0, 0.2, 0, 0]

# creating the plot
plt.pie(y, labels=cars, explode=wedge);
```

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# adding labels
cars = ['benz', 'honda', 'ford', 'nissan']

# making the wedges stand out
wedge = [0, 0.2, 0, 0]

# new colors
colrs = ['b', 'c', 'm', 'y']

# creating the plot
plt.pie(y, labels=cars, explode=wedge, colors=colrs);
```

##### Color Codes:
1. 'r' for red
2. 'g' for green
3. 'b' for blue
4. 'c' for cyan
5. 'm' for magenta
6. 'y' for yellow
7. 'k' for black
8. 'w' for white

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# adding labels
cars = ['benz', 'honda', 'ford', 'nissan']

# creating the plot
plt.pie(y, labels=cars)

# adding a legend
plt.legend()
```

#### 4. Scatter Plots
- scatter plots are used to find the relationship between different variables
- each dot shows one observation (row)
- the two variables should be of the same size (the same number of rows)

```python
# creating the dataset
x = np.array([1,2,3,4,5,6,7])
y = np.array([0,2,6,9,3,4,5])

# creating the plot
plt.scatter(x, y, c=y, cmap='RdYlGn', s=500, edgecolors='black')
plt.colorbar()
```

**Observations:**
- create the scatter plot with x and y
- c=y colors the points based on their y values
- cmap: the colormap used will be set to 'Red-Yellow-Green'
- s: dot (marker) size
- edgecolor: the color around the dots (markers)
- plt.colorbar: adds the color bar to the plot, helping to interpret the mapping of data values to the colors

```python
# creating the dataset
x = np.array([1,2,3,4,5,6,7])
y = np.array([0,2,6,9,3,4,5])

# creating the plot
plt.scatter(x, y, c=y, cmap='RdYlGn', s=500, edgecolors='black')
plt.colorbar()
plt.gray() # makes everything gray scale
```

#### 5. Histogram
- one of the most popular plots

**Types of Distribution:**

#### 6. Heatmaps

### Conclusion:
- Matplotlib allows customization of plots, including adjusting colors, markers, lines, label, and tick marks

#### Multivariate Visualization
- comparing more than two variables
- there are three popular multivariate plots:
    1. pairplot()
    2. corr() for correlation between the variables (positive and negative correlation)
    3. heatmap() shows correlation as well

```python
# 1. let's create a pairplot for numeric variables only
sns.pairplot(titanic, vars=['Age', 'Fare', 'SibSp']);
```

**Observations:**
- quickly see how each variable is a dataset relates to other variables
- helps in understanding the individual distributino and the relationships simultaniously

#### Correlation:
1. what is correlation? a relationship
2. what are the types of correlation? positive (when all move in the same direction) or negative (when things move independent of each other [spending/saving])
3. which plot is most popular in presenting correlation? heatmap

```python
#1. first let's create a df that contains only numeric variables
num_vars = titanic[['Fare', 'Age', 'SibSp', 'Parch', 'Survived', 'Pclass']]
num_vars
```

```python
#2. let's get the correlation table
num_vars.corr()
```

**Observations:**
- highest positive corr is sibsp and parch
- highest negative corr is survived and sibsp

```python
# heatmap: the most popular plot in showing correlation

# setting the size of the plot with pyplot
plt.figure(figsize=(12, 8))

# from seaborn create the heatmap
sns.heatmap(num_vars.corr(), cmap='RdYlGn', annot=True);
```

**Observations:**
- uses color to represent values in a matrix format, making it easy to identify patterns, correlations, and anomalies
- quickly spot which variable are positvely or negatively correlated

**Conclusions:**
- Techniques like box plots, violin plots, and histograms effectively summarize data distributions and key statistics, making it easier to understand large datasets at a glance.
- Count plots, point plots, and swarm plots are particularly useful for comparing frequencies and distributions across different categories, highlighting trends and variations.
- Violin plots and boxen plots provide enhanced clarity in visualizing data distributions, especially in datasets with many outliers or multimodal distributions.
- Scatter plots, pair plots, and point plots are excellent for visualizing relationships between variables, allowing users to identify correlations and potential patterns.
- KDE and swarm plots focus on visualizing the density of data points, helping to understand how values are spread out and the presence of clusters or outliers.
- Heatmaps utilize color gradients to represent data values, facilitating quick identification of patterns and correlations within complex datasets.
- Swarm plots and violin plots allow for the visualization of individual data points alongside distribution patterns, providing deeper insights into the data.
- Pair plots enable simultaneous exploration of relationships between multiple variables, helping to uncover potential correlations that might not be evident from single-variable analysis.
- Box plots and heatmaps provide not only visual insights but also statistical summaries that can inform decision-making and further analysis.
- Overall, these visualization techniques enhance the accessibility of data analysis, making it easier for users to interpret complex information and derive meaningful insights.
