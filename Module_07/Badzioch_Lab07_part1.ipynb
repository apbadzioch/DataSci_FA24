## Part 1: Types of Graphs and Plots

#### 1. Line Plots
- one of the simplest plots
- they indicate the progress of a trend over time

```python
# let's create the dataset
a = np.linspace(0, 10)

# creating the line plot with matplotlib.pyplot
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
```

**Observations:**
- add semi colon (;) at end of last line to get rid off excess data (under the hood)
- call most plots by using plt.PLOTNAME from pyplot

```python
# creating the dataset
x = np.array([0, 5])
y = np.array([0, 20])

# creating the plot
plt.plot(x, y)
```

```python
# creating the dataset
y = np.array([12,0,9,2,14])

# creating the plot
plt.plot(y, color='red', marker='*')
```

**Observations:not really needed**
- blue is the default color

```python
# creating the dataset
y = np.array([0.5, 2.0, 3.5, 4.0])

# creating the plot
plt.plot(y, color='green', linestyle='dotted') # we can have 'dashed' as the line style as well
```

```python
# creating the dataset
x = pd.Series(['sunday', 'monday', 'tuesday', 'wednesday', 'thursday'])
y = pd.Series(['running', 'tennis', 'football', 'biking', 'ping-pong'])

# creating the plot
plt.plot(x, y)

# adding more items to the plot
plt.title('days vs sports') # adding a title to the graph
plt.xlabel('days') # adding x label
plt.ylabel('sports') # adding y label

# adding a grid
plt.grid()
```

#### 2. Bar Plots
- bar plots are used to compare different categories

```python
# creating the dataset
df1 = pd.DataFrame.from_dict(
    {'year': [2016, 2017, 2018, 2022],
     'men': [300, 305, 325, 350],
     'women': [280, 301, 325, 355],
     'total': [580, 606, 650, 705]
     }
)

# creating the bar plot
plt.bar(x=df1['year'], height=df1['total'])
```

#### 3. Pie Charts
- pie charts are used to indicate the proportion over the whole

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# creating the plot
plt.pie(y);
```

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# adding labels
cars = ['benz', 'honda', 'ford', 'nissan']

# making the wedges stand out
wedge = [0, 0.2, 0, 0]

# creating the plot
plt.pie(y, labels=cars, explode=wedge);
```

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# adding labels
cars = ['benz', 'honda', 'ford', 'nissan']

# making the wedges stand out
wedge = [0, 0.2, 0, 0]

# new colors
colrs = ['b', 'c', 'm', 'y']

# creating the plot
plt.pie(y, labels=cars, explode=wedge, colors=colrs);
```

##### Color Codes:
1. 'r' for red
2. 'g' for green
3. 'b' for blue
4. 'c' for cyan
5. 'm' for magenta
6. 'y' for yellow
7. 'k' for black
8. 'w' for white

```python
# creating the dataset
y = np.array([5, 20, 10, 8])

# adding labels
cars = ['benz', 'honda', 'ford', 'nissan']

# creating the plot
plt.pie(y, labels=cars)

# adding a legend
plt.legend()
```

#### 4. Scatter Plots
- scatter plots are used to find the relationship between different variables
- each dot shows one observation (row)
- the two variables should be of the same size (the same number of rows)

```python
# creating the dataset
x = np.array([1,2,3,4,5,6,7])
y = np.array([0,2,6,9,3,4,5])

# creating the plot
plt.scatter(x, y, c=y, cmap='RdYlGn', s=500, edgecolors='black')
plt.colorbar()
```

**Observations:**
- create the scatter plot with x and y
- c=y colors the points based on their y values
- cmap: the colormap used will be set to 'Red-Yellow-Green'
- s: dot (marker) size
- edgecolor: the color around the dots (markers)
- plt.colorbar: adds the color bar to the plot, helping to interpret the mapping of data values to the colors

```python
# creating the dataset
x = np.array([1,2,3,4,5,6,7])
y = np.array([0,2,6,9,3,4,5])

# creating the plot
plt.scatter(x, y, c=y, cmap='RdYlGn', s=500, edgecolors='black')
plt.colorbar()
plt.gray() # makes everything gray scale
```

#### 5. Histogram
- one of the most popular plots
- used to find the data distribution (frequency)
- the two variables should be of the same size (the same number of rows)

**Types of Distribution:**
1. Normal (zero skewness)
2. Right-Skewed 
3. Left-Skewed

#### 6. Heatmaps
- used to find out about the correlation between various variables

### Conclusion:
- Matplotlib allows customization of plots, including adjusting colors, markers, lines, label, and tick marks
- it provides many types of visualizations (line graphs, bar charts, scatter plots, pie charts, histograms, and heatmaps)
- line charts are one of the most commonly used plots in Matplotlib, ideal for trends over time
- line charts allow plotting multiple lines on the same axes
- bar charts are great for comparing discrete or categorical data
- pie charts are used for visualizing the parts of a whole and showing percentage distributions
- pie charts allow for "exploding" slices to emphasize specific selections 
- scatter plots are best for visualizing relationships or correlations betwwen 2 continuous variables 
- Matplotlib allows for extensive customization of marker size (s), colors (c), and shapes based on additional data
- histograms are ideal for visualizing the distibution of a single continuous variable
- heatmaps are useful for visualizing data intensity across a 2D matrix
